```python
import tensorflow as tf
import numpy as np

xy = np.loadtxt('lab_05_train.txt', unpack=True, dtype='float32')
x_data = xy[0:-1]
y_data = xy[-1]

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

W = tf.Variable( tf.random_uniform([1,len(x_data)], -1.0, 1.0) )


h = tf.matmul(W, X)
hypothesis = tf.div( 1.0, 1.0 + tf.exp(-h) )
cost = -tf.reduce_mean( Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis) )

a = tf.Variable(0.1)
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

for step in xrange(2001):
    sess.run( train, feed_dict={X: x_data, Y: y_data} )
    if step % 20 == 0:
        print step, sess.run( cost, feed_dict={X: x_data, Y: y_data} ), sess.run(W)
```

```python
# test: input: [bias, hour, attendance]
test1 = sess.run( hypothesis, feed_dict={X: [ [1], [2], [2] ]} )
print test1, test1 > 0.5

test2 = sess.run( hypothesis, feed_dict={X: [ [1], [5], [5] ]} )
print test2, test2 > 0.5

test3 = sess.run( hypothesis, feed_dict={X: [ [1, 1], [4, 3], [3, 5] ]} )
print test3, test3 > 0.5
```
